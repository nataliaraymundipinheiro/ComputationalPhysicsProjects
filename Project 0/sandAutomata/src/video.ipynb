```python
import numpy as np


video_text = np.loadtxt("../video.txt")
video_text
```

```python
video_text_reshape = video_text.reshape(int(len(video_text) / 25), 25, 25)
video_text_reshape.shape
```

```python
%matplotlib notebook

import IPython.display
from IPython.display import HTML
import matplotlib.pyplot as plt
import matplotlib.animation as animation


fig, ax = plt.subplots()

ax.set_xlim((0, 24))
ax.set_ylim((0, 24))

im = ax.imshow(video_text_reshape[0,:,:])

def init():
    im.set_data(video_text_reshape[0,:,:])
    return (im,)

def animate(i):
    data_slice = video_text_reshape[i,:,:]
    im.set_data(data_slice)
    return (im,)

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)

HTML(anim.to_html5_video())
anim.save("../video.mp4")
```
