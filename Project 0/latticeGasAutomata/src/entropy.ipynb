```python
import numpy as np


# Open function to open the file "results.txt".
# Read-only mode.
results = open("../results.txt", "r")

# Reads all the lines and return them as each line
# a string element in a list.
lines = results.readlines()

# Size of matrix.
size = int(lines[0])
# Number of sweeps taken.
sweeps = int(lines[1])

# Initialize array:
states = np.zeros(shape=(sweeps, size, size))

# Use these variables as index
state = 0
row = 2
col = 0

for line in lines:
    # Make sure line is not only a newline:
    if line != '\n':
        
        for char in line:            
            # Make sure char is a digit:
            if char != ' ' and char != '\n':
                # Make sure we are in bounds of sweeps still.
                if state >= sweeps:
                    break
                
                # Update value
                states[state, row, col] = char

                # Go to the next column
                col += 1

                # If next column get out of bounds,
                # go to next row.
                if col >= size:
                    col = 0
                    row += 1

                # If next row get out of bounds,
                # go to next state.
                if row >= size:
                    row = 0
                    state += 1

# Close file
results.close()

print("Done.")
```

```python
import gzip


# Save the sizes of the compressed files.
sizes = []

# For each state we have:
for state in states:
    # Compress it:
    compressed = gzip.compress(state.tobytes())
    
    # Get its size to plot.
    sizes.append(len(compressed))

print("Done.")
```

```python
# Find equilibrium value
equilibrium = sizes[4000:]
average = sum(equilibrium) / len(equilibrium)

print(average)
```

```python
# First part of grading:

import matplotlib.pyplot as plt

# Plot the graph.
plt.xlabel("Sweeps")
plt.ylabel("Number of Compressed Bits")

plt.title("Lattice Gas Automata: Entropy vs. Sweeps")

plt.plot(sizes)
plt.axhline(y=average, color='black', linestyle='dashed')

plt.savefig("../graphs/lga.png", dpi=500)
```

```python
# Second part of grading:

count = 0

# Plot the graph.
plt.xlabel("Sweeps")
plt.ylabel("Number of Compressed Bits")

plt.title("Lattice Gas Automata: Entropy vs. Sweeps (Part %d)" % (count + 1))

start = 500 * count
end = 500 * (count + 1)

x = np.arange(start, end)
y = sizes[start:end]

plt.plot(x, y)

plt.savefig("../graphs/lga_part%d.png" % (count + 1), dpi=500)

count += 1
```
